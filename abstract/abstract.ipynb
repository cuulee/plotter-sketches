# Code

```python
def ngon(n, l = 1, p = 0):
    return Polygon([(l*cos(p+θ), l*sin(p+θ)) for θ in linspace(0, 2*pi, n+1)])

def get_zigzag(shape, k = 4):
    x, y = shape.exterior.xy
    xy = stack([x, y], axis = 1)
    m = len(x)
    i = array(list(zip(range(0, m)[::k], range(m-1, 0, -1)[::k]))).reshape(-1)
    return LineString(xy[i])

class Sketch(vsketch.Vsketch):

    def draw(self) -> None:

        self.size("a4", landscape = True)
        self.scale("cm")
        
        self.scale(2)
        
        n = 4
        shapes = []
        for i in range(n):
            for j in range(n):
                zigzag1 = get_zigzag(ngon(32)).buffer(4e-2)
                zigzag2 = zigzag1.intersection(rotate(zigzag1, 120*(norm((array([i+.5-n/2, j+.5-n/2]))))))
                zigzag2 = translate(zigzag2, 2.5*i, 2.5*j)
                zigzag2 = zigzag2.buffer(6e-2)
                if type(zigzag2) == MultiPolygon:
                    shapes += zigzag2
                else:
                    shapes.append(zigzag2)
                
        shapes = MultiPolygon(shapes)
        shadow = translate(shapes, .15, .15)
        
        self.fill(1)
        self.stroke(1)
        self.penWidth(8)
        self.geometry(shapes)
        
        self.penWidth(3)
        self.noStroke()
        self.fill(2)
        self.geometry(shadow.difference(shapes.buffer(7e-2)))
        

    def finalize(self) -> None:
        self.vpype("linemerge linesimplify reloop linesort")


vsk = Sketch()
vsk.draw()
vsk.finalize()
vsk.display()
vsk.save('prints/abstract-1.svg')
```
